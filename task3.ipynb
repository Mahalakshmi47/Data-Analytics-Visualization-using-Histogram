# Visualization using Histogram 

## Importing Libraries

```python
import seaborn as sns
import pandas as pd
```

## Creating Dataframe

```python
d=pd.read_csv('Iris.csv') 
```

```python
d.head()
```

```python
d.tail()
```

```python
d.info()
```

```python
d['Species'].unique()
```

## Histogram

```python
sns.histplot(data=d,x='SepalLengthCm')
```

```python
sns.histplot(data=d,y='PetalLengthCm',x='Id')
```

```python
sns.displot(data=d,x='SepalLengthCm')
```

## Bar Plot

```python
sns.barplot(data=d,x='SepalLengthCm',y='SepalWidthCm',hue='Species')
```

```python
sns.barplot(data=d,x='PetalLengthCm',y="Id",hue='Species')
```

## Pair Plot

```python
sns.pairplot(data=d)
```

```python
sns.pairplot(data=d,hue='Species')
```

## Line Plot

```python
sns.lineplot(data=d,x='SepalLengthCm',y='PetalLengthCm',hue='Species')
```

## Categorical Plot

```python
sns.catplot(data=d,x='SepalLengthCm',y='PetalLengthCm',hue='Species')
```

## Count Plot

```python
sns.countplot(data=d,x='SepalLengthCm')
```

```python
sns.countplot(data=d,x='PetalLengthCm',hue='Species')
```

## Box Plot

```python
sns.boxplot(data=d,x='SepalLengthCm',hue='Species',hue_order='Species')
```

```python
sns.boxplot(data=d,x='SepalLengthCm',y='SepalWidthCm',hue='Species')
```

```python
sns.boxplot(data=d,x='SepalLengthCm',y='Species',hue='Species')
```

## Violin Plot

```python
sns.violinplot(data=d,x='SepalLengthCm',y='Species',hue='Species')
```

## Scatter Plot

```python
sns.scatterplot(data=d,x='SepalLengthCm',y='PetalLengthCm',hue='Species')
```

## Heat Map

```python
sns.heatmap(d.corr(),annot=True)
```
